```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Create dataset inside code (no file needed)
data = """Age,MonthlyIncome,YearsAtCompany,JobRole,Attrition
25,4000,1,Sales Executive,Yes
30,6000,3,Developer,No
35,8000,5,Manager,No
28,4500,2,HR,Yes
40,9000,7,Developer,No
32,7000,4,Sales Executive,No
29,4800,2,HR,Yes
45,10000,10,Manager,No
"""

df = pd.read_csv(StringIO(data))

# Show data
print("DATA:")
print(df)

# Attrition count
print("\nATTRITION COUNT:")
print(df['Attrition'].value_counts())

# Graph 1
df.groupby('Attrition')['MonthlyIncome'].mean().plot(kind='bar')
plt.title("Average Salary vs Attrition")
plt.show()

# Graph 2
df.groupby('Attrition')['YearsAtCompany'].mean().plot(kind='bar')
plt.title("Years at Company vs Attrition")
plt.show()
```
